```python
import os
from dotenv import load_dotenv
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

```python
training_solutions = pd.read_excel("../../data/raw/widsdatathon2025/TRAIN/TRAINING_SOLUTIONS.xlsx")
training_solutions.head()
```

```python
# Pie Chart for ADHD Outcome
labels = training_solutions['ADHD_Outcome'].map({0: 'Other/None', 1: 'ADHD'}).value_counts().index
sizes = training_solutions['ADHD_Outcome'].value_counts().values
colors = ['#ff9999', '#66b3ff']
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Distribution of ADHD Outcome')
plt.show()
```

```python
# Pie Chart For Sex
labels = training_solutions['Sex_F'].map({0: 'Male', 1: 'Female'}).value_counts().index
sizes = training_solutions['Sex_F'].value_counts().values
colors = ['#ff9999', '#66b3ff']
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Distribution of Sex')
plt.show()
```
